## Training the Apriori model on the dataset

```python
# WARNING: Make sure to upload the apyori.py file into this Colab notebook before running this cell
from apyori import apriori
rules = apriori(transactions=transactions, min_support = 0.002, min_confidence = 0.40, min_lift = 2, min_length = 80,use_colnames=True)
```
